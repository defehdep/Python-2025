```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Гистграмма

rng = np.random.default_rng(1)
data = rng.normal(size=1000)

plt.hist(data,
        bins = 50,
        density = True,
        alpha = 0.5,
        histtype = 'step',
        edgecolor = 'red'
)
plt.show()
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(6, 2, 1000)


args = dict(
    alpha = 0.3,
    bins = 50
)
plt.hist(x1, **args)
plt.hist(x2, **args)
plt.hist(x3, **args)

plt.show()
```

```python
print(np.histogram(x1, bins=1))
print('------')
print(np.histogram(x1, bins=2))
print('------')
print(np.histogram(x1, bins=40))
```

```python
# Двумерные гистограммы

mean = [0,0]
cov = [[1,1],[1,2]]

x,y = rng.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x,y,bins=50)
cb = plt.colorbar()
cb.set_label('Points')
plt.show()

print(np.histogram2d(x,y,bins = 10))
```

```python
plt.hist2d(x,y,bins=50)
plt.hexbin(x,y,gridsize = 30) # hexbin
cb = plt.colorbar()
cb.set_label('Points')
plt.show()
```

```python
# Легенда

x = np.linspace(0,10,1000)

fig, ax = plt.subplots()

ax.plot(x, np.sin(x), label = 'Синус')
ax.plot(x, np.cos(x), label = 'Косинус')
ax.plot(x, np.cos(x) + 2)
ax.axis('equal')

ax.legend(
    frameon = True,  # рамка у легенды
    fancybox = True,
    shadow = True,   # тень
    loc = 'lower center' # расположение легенды
)
plt.show()
```

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0,10,1000)
for i in range(4):
    lines += ax.plot(
        x,
        np.sin(x - i + np.pi / 2),
        styles[i]
    )
ax.axis('equal')

ax.legend(lines[:2], ['line 1', 'line 2'], loc = 'upper right')

leg = mpl.legend.Legend(ax, lines[1:], ['line 2', 'line 3', 'line 4'], loc = 'lower left')
ax.add_artist(leg)

plt.grid()
plt.minorticks_on() 
plt.grid(which='minor', linestyle=':', linewidth=0.5)

plt.show()
```

```python
# Шкалы

x = np.linspace(0,10,1000)
y = np.sin(x) + np.cos(x[:,np.newaxis]) # [:,np.newaxis] - в другой оси
#plt.imshow(y, cmap = 'binary')
#plt.imshow(y, cmap = 'viridis')
#plt.imshow(y, cmap = 'RdBu')
#plt.imshow(y, cmap = 'rainbow')
plt.imshow(y, cmap = 'jet')
plt.clim(-1.5, 0.5)
plt.colorbar()
plt.show()
```
