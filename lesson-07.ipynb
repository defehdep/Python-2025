```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# чтобы задать координату каждому сабплоту используют:

grid = plt.GridSpec(2,3)

plt.subplot(grid[0,0])
plt.subplot(grid[1,2])

plt.show()
```

```python
# можно обознать какой график сколько места будет занимать

grid = plt.GridSpec(2,3)

plt.subplot(grid[:2,0])
plt.subplot(grid[0,1:])
plt.subplot(grid[1,1])
plt.subplot(grid[1,2])

plt.show()
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]

rng = np.random.default_rng(1)
x, y = rng.multivariate_normal(mean, cov, 3000).T

fig = plt.figure()
grid = plt.GridSpec(4, 4, hspace = 0.2, wspace = 0.2)

main_ax = fig.add_subplot(grid[0:3,1:4])

y_hist = fig.add_subplot(grid[:3,0], sharey = main_ax)
x_hist = fig.add_subplot(grid[3,1:], sharex = main_ax)

main_ax.plot(x, y, 'ok', markersize = 3, alpha = 0.2)

y_hist.hist(y, bins=30, orientation='horizontal', color='gray', alpha=0.7)
x_hist.hist(x, bins=30, color='gray', alpha=0.7)

plt.show()
```

```python
# создание стрелки

fig, ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')
                       # arrowprops для стрелочки
                       # xy = (6.28, 1) - координата конца стрелки
                       # xytext=(1,4) - коордниата текста
ax.annotate('Локальный максимум', xy = (6.28, 1), xytext=(1,4), 
            arrowprops = dict(facecolor='red'))
ax.annotate('Локальный минимум', xy = (3.14, -1), xytext=(2,-6), 
            arrowprops = dict(facecolor='red',arrowstyle = '->'))

plt.show()
```

```python
fig, ax = plt.subplots(4,4, sharex = True, sharey = True)

for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(5))
    axi.yaxis.set_major_locator(plt.MaxNLocator(4))

plt.show()
```
