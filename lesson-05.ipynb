```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
fig = plt.figure()

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.figure() 
plt.figure(figsize=(10,10)) #   ""

plt.subplot(3,1,1)
plt.plot(x, np.sin(x), label = '')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('')
plt.legend()

plt.subplot(3,1,2)
plt.plot(x, np.sin(x), '--g', label = '')
plt.plot(x, np.cos(x), ':b', label = '')
plt.xlabel('x')
plt.title('  ')
plt.legend()

plt.subplot(3,1,3)
plt.plot(x, np.sin(x), '--g', label = '')
plt.plot(x, np.cos(x), ':b', label = '')
plt.xlabel('x')
plt.title('  ')
plt.axis('equal') #     x  y
plt.legend()
plt.subplots_adjust(hspace = 0.5) #   ""

plt.show()
```

```python
# 
x = np.linspace(0,10,30)
plt.plot(x, np.sin(x), '.', color='g')
plt.plot(x, np.sin(x)+1, '>', color='g')
plt.plot(x, np.sin(x)+2, 's', color='g')
plt.plot(x, np.sin(x)+3, '^', color='g')

plt.show()
```

```python
x = np.linspace(0,10,30)
plt.plot(x, np.sin(x), '--p', 
         markersize=10, linewidth=2,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
plt.show()
```

```python
#  "scatter"    
x = np.linspace(0,10,30)
rng = np.random.default_rng(0)
colors = rng.random(30)
sizes = 100 * rng.random(30)

plt.scatter(x, np.sin(x), marker = 'o', c = colors, s = sizes)
plt.colorbar() # 

plt.show()
```

```python
#  
x = np.linspace(0,10,50)
dy = 0.4
y = np.sin(x) + dy * np.random.randn(50)
plt.figure(figsize=(10,5))

plt.errorbar(x, y, yerr = dy, fmt = '.k') # fmt = '.k' -   
plt.fill_between(x, y - dy, y + dy, color = 'r', alpha = 0.4) # 
plt.show()
```

```python
#    
def f(x,y):
    return np.cos(x)**100 + np.sin(x * y + 20) * np.sin(x)

x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X, Y = np.meshgrid(x,y)

Z = f(X,Y)

#plt.contour(X,Y,Z, cmap = 'coolwarm') #  : cmap = 'RdGy'
#plt.contourf(X,Y,Z, cmap = 'coolwarm') #  
plt.contour(X,Y,Z, cmap = 'coolwarm')
plt.imshow(Z, extent=[0,5,0,5], cmap = 'RdGy',  # extent=[0,5,0,5] -  
          interpolation = 'gaussian', # 
          aspect = 'equal', # aspect -  
          origin = 'lower') #  

# c = plt.contour(X,Y,Z)          #   
# plt.clabel(c)

plt.colorbar()
plt.show()
```
